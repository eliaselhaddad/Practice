```python
import matplotlib.pyplot as plt

def main():
  budget = {}
  while True:
    print("Menu:")
    print("1. Enter income")
    print("2. Enter expenses")
    print("3. View budget summary")
    print("4. Set budget goals")
    print("5. Exit program")
    choice = input("Enter your choice: ")
    
    if choice == "1":
      source = input("Enter the income source: ")
      amount = float(input("Enter the amount: "))
      budget[source] = amount
    elif choice == "2":
      category = input("Enter the expense category: ")
      description = input("Enter the expense description: ")
      amount = float(input("Enter the amount: "))
      budget[description] = -amount
    elif choice == "3":
      income = sum(budget.values())
      expenses = sum([expense for expense in budget.values() if expense < 0])
      net_balance = income + expenses
      print("Total income: $%.2f" % income)
      print("Total expenses: $%.2f" % abs(expenses))
      print("Net balance: $%.2f" % net_balance)
      
      # Create a pie chart to visualize the budget
      labels = "Income", "Expenses"
      sizes = [income, abs(expenses)]
      plt.pie(sizes, labels=labels)
      plt.show()
    elif choice == "4":
      goal_type = input("Enter the goal type (e.g. saving a certain amount per month): ")
      goal_amount = float(input("Enter the goal amount: "))
      budget["goal"] = goal_amount
    elif choice == "5":
      break
    else:
      print("Invalid choice. Please try again.")

main()
```
